```python
with open("../data/Wagner.txt") as f:
    text = f.read()

text = text.split("\n")
for mot in text:
    mot.replace("\t", "")
with open("../data/WagnerTEI.txt", encoding='UTF-8', mode="w") as tei:
    tei.write('<?xml version="1.0" encoding="UTF-8"?>\n<?xml-model href="http://www.tei-c.org/release/xml/tei/custom/schema/relaxng/tei_all.rng" type="application/xml" schematypens="http://relaxng.org/ns/structure/1.0"?>\n<?xml-model href="http://www.tei-c.org/release/xml/tei/custom/schema/relaxng/tei_all.rng" type="application/xml"schematypens="http://purl.oclc.org/dsdl/schematron"?>\n<TEI xmlns="http://www.tei-c.org/ns/1.0">\n')
    tei.write('<teiHeader>\n<fileDesc>\n<titleStmt>\n<title>')
    for metadata in text:
        if "TITRE" in metadata:
            asupprimer = metadata
            metadata = metadata.replace("TITRE ", "")
            tei.write(metadata)
            tei.write("</title>\n")
            asupprimer = text.index(asupprimer)
            suppr = text.pop(asupprimer)
    for metadata in text: 
        if "AUTEUR" in metadata:
            asupprimer = metadata
            tei.write("<author>")
            metadata = metadata.replace("AUTEUR ", "")
            tei.write(metadata)
            tei.write("</author>\n")
            asupprimer = text.index(asupprimer)
            suppr = text.pop(asupprimer)
    tei.write("</titleStmt><publicationStmt><publisher>wikisource</publisher></publicationStmt><sourceDesc><p>pris sur wikipedia</p></sourceDesc></fileDesc></teiHeader>\n<text>\n<front>\n<castList>\n")
caps = ["A", "Z", "E", "R", "T", "U", "Y", "I", "O", "P", "Q", "S", "D", "F", "G", "H", "J", "K", "L", "M", "W", "X", "C", "V", "B", "N", "É", "Ê", "Ë", "È", "À", "Ö"]

with open("../data/WagnerTEI.txt", encoding='UTF-8', mode="a") as tei:
    indexCasting = text.index("PERSONNAGES")
    indexsuppr = indexCasting
    indexCasting += 1
    for casting in text[indexCasting:-1]:
        if casting == "":
            break
        elif casting[0] in caps:
            if casting[1] in caps:
                tei.write("<castItem>")
                tei.write(casting)
                tei.write("</castItem>\n")
                indexcast = text.index(casting)
            elif casting[1] not in caps:
                tei.write("</castList>\n<set>\n<p>")
                tei.write(casting)
                tei.write("</p>\n</set>")
                indexcast = text.index(casting)
            else:
                continue
        else:
            break
    indexcast += 1
    indexcast += 1
    del text[indexsuppr:indexcast]
    tei.write("\n</front>\n<body>")
    
numerotation = 1
with open("../data/WagnerTEI.txt", encoding='UTF-8', mode="a") as tei:
    for vers in text:
        if vers == "":
            continue
        elif "ACTE" in vers and "FIN" not in vers:
            asupprimer = vers
            asupprimer = asupprimer.replace("ACTE ", "")
            if asupprimer == "PREMIER" or asupprimer == "1ER" or asupprimer == "I":
                indexmeta_acte = 1
            elif asupprimer == "DEUXIÈME" or asupprimer == "2e" or asupprimer == "2E" or asupprimer == "DEUXIEME" or asupprimer == "II":
                indexmeta_acte = 2
            elif asupprimer == "TROISIÈME" or asupprimer == "3e"or asupprimer == "3E" or asupprimer == "TROISIEME" or asupprimer == "III":
                indexmeta_acte = 3
            elif asupprimer == "QUATRIÈME" or asupprimer == "4e"or asupprimer == "4E" or asupprimer == "QUATRIEME" or asupprimer == "IV":
                indexmeta_acte = 4
            elif asupprimer == "CINQUIÈME" or asupprimer == "5e"or asupprimer == "5E" or asupprimer == "CINQUIEME" or asupprimer == "V":
                indexmeta_acte = 5
            else:
                indexmeta_acte = int(asupprimer)
            tei.write('<div type="act" n="')
            tei.write(str(indexmeta_acte))
            tei.write('">\n<head>')
            tei.write(vers)
            tei.write("</head>")
        elif "ACTE" in vers and "FIN" in vers:
            tei.write("\n</sp></div>\n</div>\n")
        elif vers[0] in caps and vers[1] in caps  and "ACTE" not in vers and "SCENE" not in vers and "SCÈNE" not in vers:
            indexvers = text.index(vers)
            indexvers = int(indexvers)-1
            if ("ACTE" in text[indexvers] or "SCÈNE" in text[indexvers] or "SCENE" in text[indexvers] or "ACT" in text[indexvers]):
                tei.write("<stage><p>")
                tei.write(vers)
                tei.write("</p></stage>")
            elif vers[-2] in caps and "ACTE" not in text[indexvers] and "SCÈNE" not in text[indexvers] and "SCENE" not in text[indexvers] and "ACT"  not in text[indexvers]:
                tei.write("</sp><sp><speaker>")
                tei.write(vers)
                tei.write("</speaker>")
        elif "(" in vers or ")" in vers:
            if "(" in vers and ")" not in vers:
                tei.write("<stage><p>")
                tei.write(vers)
            elif ")" and "(" not in vers:
                tei.write(vers)
                tei.write("</p></stage>")
            else:
                tei.write("<stage><p>")
                tei.write(vers)
                tei.write("</p></stage>")
        elif "SCENE" in vers or "SCÈNE" in vers:
            if "1" in vers or "PREMIERE" in vers or "1re" in vers or "1RE" in vers or "1E" in vers or "1e" in vers or "PREMIÈRE" in vers:
                tei.write('\n<div type ="scene" n="1">\n<head>')
                tei.write(vers)
                tei.write("</head>\n")
            else:
                tei.write('\n</sp>\n</div>\n<div type="scene" n="')
                if "II" in vers:
                    tei.write("2")
                elif "III" in vers:
                    tei.write("3")
                elif "IV" in vers:
                    tei.write("4")
                elif "V" in vers:
                    tei.write("5")
                elif "VI" in vers:
                    tei.write("6")
                elif "VII" in vers:
                    tei.write("7")
                elif "VIII" in vers:
                    tei.write("8")
                elif "IX" in vers:
                    tei.write("9")
                elif "X" in vers:
                    tei.write("10")
                tei.write('"><head>')
                tei.write(vers)
                tei.write("</head>\n")
        elif "Situation :" in vers:
            situation = vers
            situation = situation.replace("Situation : ", "")
            tei.write("\n<stage><p>")
            tei.write(situation)
            tei.write("</p></stage>")
        else:
            tei.write('\n<l n="')
            tei.write(repr(numerotation))
            tei.write('">')
            tei.write(vers)
            tei.write("</l>")
            numerotation += 1
    tei.write("\n</div>\n</body>\n</text>\n</TEI>")
    
    
    
```
